```python
!pip install paddlepaddle paddleocr

from paddleocr import PaddleOCR, draw_ocr
from PIL import Image
import fitz  # PyMuPDF
import numpy as np # import numpy

ocr = PaddleOCR(use_angle_cls=True, lang="en")

pdf_path = "/content/drive/MyDrive/Colab Notebooks/Ligare Sample Data/sample_data/driverlicenses.pdf"

doc = fitz.open(pdf_path)

# Define keywords to look for
keywords = ["driver", "license", "identification", "id", "dob", "date of birth", "expiration", "class", "endorsements", "chauffeur-license", "driver-license", "chauffeur license", "identification card", "commercial driver license", "driver license*", "driver's", "operator's"]

for page_num in range(doc.page_count):
    page = doc[page_num]
    pix = page.get_pixmap()
    img = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)
    # Convert the PIL Image to a NumPy array
    img = np.array(img)
    result = ocr.ocr(img, cls=True)

    # Check if result is not None and contains detected text
    if result and result[0]:
        print(f"Page {page_num + 1}:")
        words_on_page = []
        for line in result:
            for word_info in line:
                words_on_page.append(word_info[1][0].lower()) # Convert to lowercase for case-insensitive matching

        # Check for keywords, including individual words in words_on_page
        is_driver_license = any(keyword in word.lower() for keyword in keywords for word in words_on_page)

        if is_driver_license:
            print("Is a driver's license.")
        else:
            print("NOT a driver's license.")

        print(words_on_page)  # Print words for debugging
        print("-" * 20)
    else:
        print(f"Page {page_num + 1}: No text detected")
        print("-" * 20)
```
